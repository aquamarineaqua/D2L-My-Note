```python
import tensorflow as tf
import torch
torch.set_printoptions(precision=8)
import numpy as np
```

## 数据索引、数据收集

```python
data = np.random.uniform(size=[2,10,4])
data_2 = np.random.uniform(size=[3,3,4])
data_3 = np.random.uniform(size=[2,3,4])
data
```

### 根据索引号收集数据（tensorflow）
我们指导切片只能提取规则数据，gather方法则能用于提取不规则非连续数据<br>
tf.gather(x, 索引方案, axis)

```python
# tensorflow

# 设有2个班级，每班10名学生，每名学生4门成绩：
b = tf.constant(data)
print(b)
print('*'*40)

print('取第一个班的成绩册：', tf.gather(b, [0], axis=0))
print('抽取学号为1、3、5、7、9号同学的成绩：', tf.gather(b, [0,2,4,6,8], axis=1))
print('*'*40)

b2 = tf.gather(b, [1], axis=0)
b3 = tf.gather(b2, [0,2,7], axis=1)
b4 = tf.gather(b3, [0,3], axis=2)
print('抽取第2个班级，第1、3、8名同学，第1、4门成绩：', b4)
print('抽取第2个班级，第1名同学，第1门成绩 和 第1个班级，第2名同学，第4门成绩：', tf.gather_nd(b, [[1,0,0],[0,1,3]]))  # gather_nd中，第二个参数传入维度索引方案，可为嵌套列表。
```

### 根据索引号收集数据（pytorch）

```python
# 设有2个班级，每班10名学生，每名学生4门成绩：
a = torch.tensor(data)
print(a)
print('*'*40)

# pytorch的gather函数十分的反人类：
index1 = torch.zeros((1,10,4), dtype=torch.long)
print('取第一个班的成绩册：', torch.gather(a, dim=0, index=index1))  # pytorch的gather函数，传入的index必须为tensor
# 这是一个三维空间中的索引。关于gather的用法，参见：https://blog.csdn.net/takedachia/article/details/124027651
index2 = torch.tensor([0,2,4,6,8])  # 在dim=1维上，需要取[0,2,4,6,8]这几个元素
index2 = torch.stack([index2 for _ in range(4)], dim=-1)  # 往后一维堆叠
index2 = torch.stack([index2 for _ in range(2)], dim=0)   # 往前一维堆叠
print(index2.shape)  # 形状应为（2, 5, 4）
print('抽取学号为1、3、5、7、9号同学的成绩：', torch.gather(a, dim=1, index=index2))
print('*'*40)

# 我们使用： torch.index_select(input, dim, index, out=None) → Tensor
# 注意，传入的index必须为tensor
index1 = torch.tensor([0])
print('取第一个班的成绩册：', torch.index_select(a, dim=0, index=index1))
index2 = torch.tensor([0,2,4,6,8])
print('抽取学号为1、3、5、7、9号同学的成绩：', torch.index_select(a, dim=1, index=index2))
print('*'*40)

a2 = torch.index_select(a, dim=0, index=torch.tensor([1]))
a3 = torch.index_select(a2, dim=1, index=torch.tensor([0,2,7]))
a4 = torch.index_select(a3, dim=2, index=torch.tensor([0,3]))
print('抽取第2个班级，第1、3、8名同学，第1、4门成绩：', a4)
# 抽取第2个班级，第1名同学，第1门成绩 和 第1个班级，第2名同学，第4门成绩。这个不如直接用切片索引来的方便。
```

### 根据掩码（布尔索引）收集数据（tensorflow）

```python
# tf.boolean_mask(x, mask, axis)
# mask可为简单列表形式，也可以是嵌套列表的矩阵

b = tf.constant(data_2)  # 设有3个班级，每班3名学生，每名学生4门成绩
print('提取第1、3个班级的数据：', tf.boolean_mask(b, [True, False, True], axis=0))
print()

b = tf.constant(data_3)
print('提取第1个班级，第1，2名同学的成绩数据；提取第2个班级，第2，3名同学的成绩数据：', 
        tf.boolean_mask(b, [[True, True, False], [False, True, True]]))

print('提取第1个班级，第2名同学的第3门成绩数据', 
        tf.boolean_mask(b, [[[False, False, False, False],
                            [False, False, True, False],
                            [False, False, False, False]],
                            [[False, False, False, False],
                            [False, False, False, False],
                            [False, False, False, False]]]))
```

### 根据掩码（布尔索引）收集数据（pytorch）

```python
# torch.masked_select(input, mask, out=None) → Tensor
# mask是一个与原tensor同形状的tensor

a = torch.tensor(data_2)  # 设有3个班级，每班3名学生，每名学生4门成绩
# 由于pytorch的masked_select函数只能传入全体索引，不能指定维度，所以索引张量要自己扩个容。
# 同时，返回的tensor是一个拉伸成一维的tensor。
index1 = torch.tensor([True, False, True])
index1 = torch.stack([index1 for _ in range(3)], dim=-1)
index1 = torch.stack([index1 for _ in range(4)], dim=-1)
print(index1.shape == a.shape)
print('提取第1、3个班级的数据：', torch.masked_select(a, index1))
# 所以，pytorch的masked_select方法并不好用。

a = torch.tensor(data_3)
print('提取第1个班级，第2名同学的第3门成绩数据', 
    torch.masked_select(a, torch.tensor([[[False, False, False, False],
                                        [False, False, True, False],
                                        [False, False, False, False]],
                                        [[False, False, False, False],
                                        [False, False, False, False],
                                        [False, False, False, False]]])))
```

### 根据条件收集数据（tensorflow）

```python
# 根据条件收集数据    tensorflow
# tf.where(cond, a, b)
# cond是布尔索引
# 如果 cond[i]=True, x[i]=a[i]
# 如果 cond[i]=False, x[i]=b[i]
# 当a, b无确定值时，返回cond中所有True元素的索引坐标。

b1 = tf.ones([3,3])
b2 = tf.zeros([3,3])
cond = [[True,False,False], [False,True,False], [False,False,True]]
print(tf.where(cond, b1, b2))
```

### 根据条件收集数据（pytorch）

```python
# 这里基本同tensorflow

a1 = torch.ones([3,3])
a2 = torch.zeros([3,3])
cond = torch.tensor([[True,False,False], [False,True,False], [False,False,True]])
print(torch.where(cond, a1, a2))
```
